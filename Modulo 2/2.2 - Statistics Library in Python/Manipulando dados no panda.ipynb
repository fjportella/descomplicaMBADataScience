```python
import pandas  as pd
#Biblioteca para buscar dados das ações
import yfinance as yf
```

```python
acao = yf.Ticker('PETR4.SA')
data = acao.history(period='1y')
data #lista as informações do dataframe
```

```python
df = data.filter(items=['Open'])
df
```

```python
#Filtrando colunas do dataframe
df = data.filter(items=['Open', 'Close'])
df
```

```python
#axis=1 é o eixo, que neste caso são as colunas. Então irá procurar todas as colunas as palavras que
# tenham as letras gh. Irá retornar High porque tem as letras gh. Filter é para as colunas
# Não serve para procurar no conteúdo.
df = data.filter(like='gh', axis=1)
df
```

```python
# Neste caso é possível fazer o filtro dentro dos valores, como por exemplo um filtro por data
# passa o dataframe, depois o dataframe e o nome da coluna a ser filtrada
df = data[(data.index > '2020-10-01')]
df
```

```python
#Selecionao uma coluna desejada
#df = data.Close
df = data['Close']
df
```

```python
#Selecionando mais de uma coluna (usa 2 colchetes)
df = data[['Open', 'Close', 'High']]
df
```

```python
#Retorna os dados do campo indice, que neste caso é campo com as datas
df = data.index
df
```

```python
#Excluindo linhas usa o drop. Usa-se o indice axis, que no caso de linhas é zero.
# Neste caso, serão excluídas as 2 últimas linhas
df = data.drop(['2024-03-14', '2024-03-15'], axis=0)
```

```python
df
```

```python
#Excluindo linhas usa o drop. Usa-se o indice axis, que no caso de colunas é 1.
# Neste caso, serão excluídas as colunas High e Low
df = data.drop(['High', 'Low'], axis=1)
df
```

```python
#Fazendo substituições dentro do dataframe, usando Map
#Criando nova coluna new_dividends para manipular o valor da coluna Dividends
#Pesquisando onde tiver valor 0 na coluna Dividens, criando valor 2 na nova coluna
df['new_dividends'] = data['Dividends'].map({0:2.00})
df
```

```python
#Fazer alterações com uma condição usando função 'map'
df['new_dividends'] = data['Dividends'].map({0:2.00}).where(data.index == '2024-03-12')
df
```

```python
#Verificando se há dados faltantes, que pode estragar uma análise
# o camando é IS NULL
#Criando um dicionário com uma lista para A, B e C

df_missing = pd.DataFrame({'A' : [0,1,2,None,4],
                           'B' : [5,6,None, 8,9],
                           'C' : ['a', 'b', 'a', 'a', None]})
df_missing
```

```python
#retorna valores nulos. mas tras Falso e Verdadeiro para cada linha e coluna, o que dificulta visualização 
df_missing.isnull()
```

```python
# retornando valores nulos, usando sum, que indica quantos nulos há em cada coluna
df_missing.isnull().sum()
```

```python
#Para solucionar esse caso da coluna faltante, uma das alternativas é fazer uma média para os faltantes
#mas, tomando cuidado para não distorcer as informações. outra forma é usar a moda (que é o que mais repete)
#para preencher com a média, usar fillna

media = df_missing['A'].mean() #achando a média
#moda = df_missing['A'].mode() #achando a moda
df_missing = df_missing['A'].fillna(media)
df_missing
```
