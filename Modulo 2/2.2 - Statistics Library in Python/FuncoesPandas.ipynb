```python
#criando função própria
def duplica_bolsa(rows):
    nova_bolsa = rows * 2
    return nova_bolsa
```

```python
#criando um novo dataframe com alguns campos do dataframe original
df = df_prouni[['uf_busca', 'curso_busca', 'mensalidade', 'bolsa_integral_ampla']]
df
```

```python
#usando a função criada usando apply.
#criando o campo nova_bolsa aplicando a função no campo bolsa_integral_ampla, duplicando os valores.
df['nova_bolsa'] = df['bolsa_integral_ampla'].apply(duplica_bolsa)
```

```python
df
```

```python
#trabalhando campos que retornaram NAN. para quem for NAN será colocado valor igual a 1
df['nova_bolsa'] = df['nova_bolsa'].fillna(1)
df
```

```python
#FUNÇÕES LAMBDA (funções anônima)
#ao invés de usar a função criada acima, a função lambda permite fazer a mesma função em uma única linha
#neste caso, lambda seria igual def, x seria o return e x * 2 seria o cálculo da outra função
#vale a pena quando o cálculo é simples
df['nova_bolsa_lambda'] = df['bolsa_integral_ampla'].apply(lambda x: x * 2)
df
```

```python
#lambda usando assign.
#usa o assign para criar um novo campo chamado nova_bolsa_lambda2, usando a função anônima lambda,
#pegando o campo bolsa_integral_ampla multiplicando por 2
df = df.assign(nova_bolsa_lambda2 = lambda x : (x['bolsa_integral_ampla'] * 2))
df
```
