Usando a Biblioteca RE do Python

```python
#para usar expressões regulares no python usa-se a biblioteca RE
import re
```

```python
#procurando antonio
match = re.match('Antonio', 'Rua Antonio Basilio')
print(match)
```

```python
#procurando antonio
search = re.search('Antonio', 'Rua Antonio Basilio')
print(search)
```

```python
#procurando tudo que for div
findall = re.findall('div', '<div class=container > <div class=barra>')
print(findall)
```

```python
#divide a string em uma lista de acordo com o caracter que quer que faça a separação
split = re.split('c', 'Descomplica')
print(split)
```

```python
#substitui uma string por outra de acordo com a busca
sub = re.sub('uma plataforma', 'a melhor plataforma', 'A Descomplica é uma plataforma de ensino online')
print(sub)
```

```python
#compila um padrão de expressão regular para ser usada em outras partes do código
#no caso pediu para buscar todos os numeros em uma string
p = re.compile(r'[0-9]+')
p.findall('12 campos de futebol, 10 bolas, 25 chuteira')
```

Expressões Regulares Usando Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('/workspaces/descomplicaMBADataScience/Bovespa.csv')
df
```

```python
df_vale = df[df['Ticker'] == 'VALE3'].copy()
df_vale
```

```python
df_vale['Ticker_Name'] = 'VALE SA'
df_vale
```

```python
#busca todos os campos que começam com Ticker
reg = df_vale.filter(regex='^Ticker')
reg
```

```python
#todos os campos com todos que terminem com 'se'
reg = df_vale.filter(regex='se$')
reg
```

```python
#todos os campos com todos que tenham '_'
reg = df_vale.filter(regex='.[_].')
reg
```

```python
#substitui tudo que terminar com 3 para 03
df_vale['Ticker'] = df_vale['Ticker'].str.replace('.3', '03', regex=True)
df_vale
```
