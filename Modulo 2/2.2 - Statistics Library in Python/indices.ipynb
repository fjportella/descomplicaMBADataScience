LOC

```python
import pandas as pd
```

```python
df_acoes = pd.read_csv('/workspaces/descomplicaMBADataScience/Bovespa.csv')
df_acoes
```

```python
df_vale = df_acoes[df_acoes['Ticker'] == 'VALE3']
df_vale
```

```python
#LOC é um comando que permite pegar uma determinada coluna pelo nome dela
df1 = df_vale.loc[:,'Date']
df1
```

```python
df2 = df_vale.loc[257:261, 'Date']
df2
```

```python
df3 = df_vale.loc[257:261, ['Date', 'Open', 'High']]
df3
```

ILOC

```python
#ILOC - diferente do LOC que retorna pelo nome da coluna, ILOC retorna pelos indices das colunas
#neste caso, pega todos os valores da coluna 1
df1 = df_vale.iloc[:, 1]
df1
```

```python
#Acessando primeira linha do dataframe usando iloc
df2 = df_vale.iloc[0]
df2
```

```python
#retorna 1 linhas, retornando todas as colunas
df3 = df_vale.iloc[0:1, :]
df3
```

```python
#retorna tudas as linhas e duas colunas (coluna 2 até 3)
df4 = df_vale.iloc[:, 2:4]
df4
```

```python
#trazendo das colunas 1 a 3 (3 é exclusivo) e das linhas 2 a 4 (4 é exclusivo)
df5 = df_vale.iloc[1:3, 2:4]
df5
```

RESET INDICES

```python
# Ao aplicar o filtro, ao invés de voltar os índices originais, colocando os indices iniciando do 0
# vai criar uma nova coluna com os novos índices, e ao lado irá criar uma coluna com o nome index com os indices antereiores
df_vale.reset_index()
```

```python
#para retirar a coluna dos indices anteriores. cria e deleta a coluna indice
df_vale.reset_index(drop=True)
```

```python
df_vale.loc[1:10, 'Date']
```

```python
# quando foram dados os comandos para resetar o indice, não fez no dataframe orignal, e sim criou uma cópia
# mantendo os dados originais. para fazer isso, faremos usando inplace e drop para excluir a coluna indice original
df_vale.reset_index(inplace=True, drop=True)
df_vale
```

```python
# Exemplo de modificação de indice sem seguir a sequencia. para isso criamos os dados de uma empresa
empresa = {'email': ['jose@email.com', 'fabricio@email.com'],
           'departamento': ['marketing', 'ti'],
           'nome': ['José Silva', 'Fabricio Mattos']}

df_empresa = pd.DataFrame(empresa)
df_empresa
```

```python
#trocando o indice para que o indice não seja mais numérico e sim o campo email
df_empresa.set_index('email')
```

```python
#contudo, mesmo fazendo a mudança acima, não altera o original, e sim é criado uma cópia
df_empresa
```

```python
#para efetivar então precisamos usar o parametro inplace
df_empresa.set_index('email', inplace=True)
df_empresa
```
