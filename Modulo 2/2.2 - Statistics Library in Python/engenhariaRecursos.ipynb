```python
import pandas as pd
```

```python
df = pd.read_csv('/workspaces/descomplicaMBADataScience/Bovespa.csv')
df
```

```python
df_vale = df[df['Ticker'] == 'VALE3']
df_vale
```

```python
df_vale.dtypes
```

```python
#converte campo data que era objeto para tipo data
df_vale['Date'] = pd.to_datetime(df_vale['Date'], format='%d/%m/%Y')
```

```python
#converte campo open que era objeto para tipo float
# Passo 1: Substituir vÃ­rgula por ponto no campo 'Open'
df_vale['Open'] = df_vale['Open'].str.replace(',', '.')

# Passo 2: Converter o campo 'Open' para float
df_vale['Open'] = df_vale['Open'].astype(float)
```

```python
df_vale['High'] = df_vale['High'].str.replace(',', '.')
df_vale['High'] = df_vale['High'].astype(float)

df_vale['Low'] = df_vale['Low'].str.replace(',', '.')
df_vale['Low'] = df_vale['Low'].astype(float)

df_vale['Close'] = df_vale['Close'].str.replace(',', '.')
df_vale['Close'] = df_vale['Close'].astype(float)
```

```python
df_vale.dtypes
```

```python
#criando campo mÃ©dia movel para 5 dias e 21 dias
df_vale['mm5d'] = df_vale['Close'].rolling(5).mean()
df_vale['mm21d'] = df_vale['Close'].rolling(21).mean()
```

```python
#criando outro dataframe excluindo algumas colunas que nÃ£o interessam. passando axis=1 para indicar que serÃ¡ por coluna
featuares = df_vale.drop(['Date', 'Ticker'], axis=1)
featuares
```

```python
#Dataset Peso e Altura
df_altura = pd.read_csv('/workspaces/descomplicaMBADataScience/weight-height.csv')
df_altura
```

```python
from matplotlib import pyplot as plt
```

```python
#grÃ¡fico histograma: weight Ã© o campo, bins quantidade de barras, rwidth que Ã© largura da barra
plt.hist(df_altura.Weight, bins=30, rwidth=0.8)
```

```python
#calcurando o zscore, que Ã© a quantidade de desvios padrÃ£o. (mean Ã© mÃ©dia e std que Ã© o desvio padrÃ£o)
df_altura['zscore'] = (df_altura.Weight - df_altura.Weight.mean()) / df_altura.Weight.std()
df_altura
```

```python
#verifica se hÃ¡ 2 zscore ou seja, 2 desvio padrÃ£o (maior ou menor que 2) (| significa ou)
df_out2 = df_altura[(df_altura.zscore < -2) | (df_altura.zscore > 2)]
df_out2
```

```python
#verifica se hÃ¡ 3 zscore ou seja, 3 desvio padrÃ£o (maior ou menor que 3) (| significa ou)
df_out3 = df_altura[(df_altura.zscore < -3) | (df_altura.zscore > 3)]
df_out3
```
